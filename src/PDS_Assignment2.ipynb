#### From above Min, Max, Median for Mileage and Power columns we can conclude Mileage is a  Normal Distribution.
#### Power is a Skewed Distribution. Both are Numerical type.
### For Mileage we use Mean. It is a Normal Distribution.
### For Power we use Median. It is a Skewed Data. It has Outliers. Mean and Max difference is too high.

```python
df['Mileage'].fillna(df.Mileage.astype(float).mean(),inplace=True)
```

```python
df['Power'].fillna(df.Power.astype(float).median(),inplace=True)
```
